```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import random
import json
```

```python
dataset_root = Path('/root/data/processed/drive_and_act/')
views = list(dataset_root.iterdir())
num_col = 2
fig, axes = plt.subplots(len(views), num_col, figsize=(15, 15))

for id, view in enumerate(views):
    for col in range(num_col):
        img = random.choice(list((dataset_root / view / 'images' / 'train').iterdir()))
        axes[id][col].imshow(plt.imread(str(img)))
        axes[id][col].set_title(view.name)
        # with (dataset_root / view / 'annotations' / 'keypoint_detection_results' / 'keypoint_detection_train.json').open() as f:
        # kp = np.array(image_id_mapper[sample['id']]['keypoints']).reshape([-1, 3])[:,:2]
        # axes[idx][i].scatter(kp[:13, 0], kp[:13, 1], c=joint_colors[:13])
        axes[id][col].axis('off')
fig.tight_layout()
fig.savefig('./img_output/drive_and_act_sample_img.png')
```
