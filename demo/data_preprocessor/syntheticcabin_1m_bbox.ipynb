```python
# count = len(processed_files)
count = 0
count_all_files = len(all_files)
stats = []
# print(f'processed: {count} / {count_all_files} = {(count / count_all_files):.4f}.')
len(remaining_files)
# with annotation_folder.open() as f:
for file_name in iter(all_files):
    with (raw_annotation_folder / file_name).open() as i:
        info = json.load(i)
        img_name = info['img_name']
        img_id = img_name.split('.')[0].split('_')[1]
        item = {
            'name': info['humans'][0]['name'],
            'camera_pose': info['camera_parameters']['name'],
            'environment_name': info['experiments_meta_data']['environment_name']
        }
        stats.append(item)
        count += 1
    if count % 10000 == 0:
        print(f'processed: {count} / {count_all_files} = {(count / count_all_files):.4f}.')
print('finish')
```

```python
stat_df = pd.DataFrame.from_records(stats)
```
