```python
import pandas as pd
import json
from experiments.experiment_config import (
    ALL_LIGHTNING_MODELS,
    SYNTHETIC_CABIN_VIEWPOINTS
)
model_names = [model.__name__ for model in ALL_LIGHTNING_MODELS]
```

```python
all_results = {}

for view_point in SYNTHETIC_CABIN_VIEWPOINTS:
    results = []
    result_indices = []
    for model_name in model_names:
        model_path = f'saved_lifter_2d_3d_model/rq1/{model_name}/{view_point}'
        with open(f'{model_path}/test_result.json') as f:
            raw_result = json.loads(f.read())
            result = {
                **raw_result['pjpe'][0],
                'MPJPE': raw_result['mpjpe']
            }
            results.append(pd.DataFrame(result, index=[model_name]))
    all_results[view_point] = pd.concat(results)
    all_results[view_point].loc['AVERAGE'] = all_results[view_point].mean()
```

```python
for view_point in all_results.keys():
    print(f'{view_point=}')
    display(all_results[view_point])
```
