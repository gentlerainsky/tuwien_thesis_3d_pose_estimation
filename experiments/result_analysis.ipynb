```python
import json
import pandas as pd
```

```python
result_file_1 = 'saved_lifter_2d_3d_model/LitSimpleBaselineLinear/drive_and_act/a_column_co_driver/all_actors/test_result.json'
result_file_2 = 'saved_lifter_2d_3d_model/LitSimpleBaselineLinear/drive_and_act/a_column_co_driver/all_actors_all_activities/test_result.json'
```

```python
with open(result_file_1) as f:
    result1 = json.loads(f.read())

with open(result_file_2) as f:
    result2= json.loads(f.read())
```

```python
df1 = pd.DataFrame(result1['activity_mpjpe'], index=[1])
df2 = pd.DataFrame(result2['activity_mpjpe'], index=[1])
df1 = df1.T[df1.T.index!='sitting_still']
df2 = df2.T[df2.T.index!='sitting_still']
pd.concat([df1, df2, (df2-df1)], axis=1)
```

```python
df1 = pd.DataFrame(result1['pjpe'], index=[1])
df2 = pd.DataFrame(result2['pjpe'], index=[1])
df1 = df1.T[df1.T.index!='sitting_still']
df2 = df2.T[df2.T.index!='sitting_still']
pd.concat([df1, df2, (df2-df1)], axis=1)
```
